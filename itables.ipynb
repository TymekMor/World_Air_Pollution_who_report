# 1. ITables 2.0

```python
import pandas as pd
# Wczytanie pliku Excel
df = pd.read_excel("who_aap_2021_v9_11august2022.xlsx", sheet_name="AAP_2022_city_v9")

# Przygotowanie nagłówków kolumn
df.columns = df.columns.str.strip().str.replace(" ", "_").str.replace("(", "").str.replace(")", "").str.replace("/", "_").str.lower()
```

Przy tak dużym zbiorze danych, gdzie analizujemy cały świat a analiza ma charakter zbiorczy. Wykorzystanie bibltioteki ITables jest dobrym narzędziem, tylko do wstępnej inspekcji danych. Przy dalszych etapach analizy nie będzie wnosić większej wartości. Ta biblioteka, byłaby bardziej przydatna na mniejszych zbiorach danych gdzie analiza ma charakter bardziej szczegółowy.

```python
from itables import show
import itables

# Konfiguracja opcji ITables
itables.options.lengthMenu = [10, 25, 50, 100]
itables.options.columnDefs = [{"width": "200px", "targets": "_all"}]

# Wyświetlenie interaktywnej tabeli
show(df)
```
